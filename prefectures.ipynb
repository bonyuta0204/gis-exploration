```python
import geopandas as gpd
```

```python
# 行政区域情報
prefectures = gpd.read_file("/content/drive/MyDrive/GIS/N03-20240101_GML/N03-20240101_prefecture.shp")
prefectures.to_crs(epsg=4326,inplace=True)
prefecture_polygons = prefectures.dissolve(by='N03_007')
```

```python
simplified_prefectures = prefecture_polygons.copy()
simplified_prefectures['geometry'] = simplified_prefectures['geometry'].simplify(tolerance=0.01)

simplified_prefectures.columns
```

```python
simplified_prefectures.to_file("/content/drive/MyDrive/GIS/prefectures/simple_polygon.shp", encoding='utf-8')
```
